## 📊 CELL 3 — Mark Analysis: What You Need in Finals

```python
def analyze_marks():
    """Calculate finals needed, risk level, and urgency score per subject."""
    results = []
    for code, info in INTERNALS.items():
        exam_info   = EXAMS[code]
        internal    = info['internal']
        complexity  = info['complexity']
        finals_needed = APLUS_THRESHOLD - internal          # out of 50
        finals_pct    = (finals_needed / 50) * 100
        days_left     = (exam_info['date'] - TODAY).days
        pressure      = finals_needed / 50
        urgency_score = round(pressure * complexity * 10, 1)

        if finals_needed > 45:   risk = '🔴 HIGH RISK'
        elif finals_needed > 35: risk = '🟠 MEDIUM'
        else:                    risk = '🟢 SAFE'

        results.append({
            'code':           code,
            'name':           info['name'],
            'internal':       internal,
            'finals_needed':  round(finals_needed, 2),
            'finals_pct':     round(finals_pct, 1),
            'days_left':      days_left,
            'exam_date':      exam_info['date'].strftime('%d %b'),
            'complexity':     complexity,
            'urgency_score':  urgency_score,
            'risk':           risk,
        })
    return sorted(results, key=lambda x: -x['urgency_score'])

data = analyze_marks()

# ─── Pretty table ───
headers = ['Code', 'Subject', 'Internal/50', 'Finals Needed/50', 'Finals %', 'Exam Date', 'Days Left', 'Urgency', 'Risk']
rows = [
    [d['code'], d['name'], d['internal'], d['finals_needed'], f"{d['finals_pct']}%",
     d['exam_date'], d['days_left'], d['urgency_score'], d['risk']]
    for d in data
]

display(HTML(h('📊 Mark Analysis — Finals Required for A+ (85/100)')))
print(tabulate(rows, headers=headers, tablefmt='fancy_grid', numalign='center'))

display(HTML("""
<div style='background:#fff3cd;border-left:4px solid #ffc107;padding:12px 16px;margin:10px 0;border-radius:4px'>
<b>⚡ Formula:</b> Total = Internal(50) + Final(50) = 100. A+ = 85+. Finals needed = 85 − Internal.<br>
<b>Urgency Score</b> = (Finals Needed / 50) × Complexity × 10. Higher = study more.
</div>
"""))
```

## 🏆 CELL 4 — Priority Ranking & Hours Budget

```python
def compute_hours_budget():
    """Compute available study hours by day type up to last exam."""
    total_hrs = 0
    wkday_count = sat_count = sun_count = 0
    last_exam = datetime.date(2026, 6, 20)
    d = TODAY
    while d <= last_exam:
        dow = d.weekday()   # 0=Mon … 6=Sun
        is_exam_day = any(e['date'] == d for e in EXAMS.values())
        if dow == 5:          # Saturday
            hrs = STUDY_HOURS_SAT; sat_count += 1
        elif dow == 6:        # Sunday
            hrs = STUDY_HOURS_SUN; sun_count += 1
        elif is_exam_day:
            hrs = 1.5         # morning quick review only
        else:
            hrs = STUDY_HOURS_WKDAY; wkday_count += 1
        total_hrs += hrs
        d += datetime.timedelta(days=1)
    return total_hrs, wkday_count, sat_count, sun_count

total_hrs, wk, sat, sun = compute_hours_budget()

display(HTML(h('🏆 Priority Ranking & Hours Budget')))

# ─── Priority table ───
rank_headers = ['Rank', 'Subject', 'Urgency Score', 'Risk', 'Recommended Hours', 'Strategy']
rank_rows = []
hour_weights = [0.22, 0.20, 0.16, 0.14, 0.11, 0.09, 0.08]   # sum = 1.0
strategies = [
    'Past papers + architecture diagrams from memory daily',
    'All NLP algorithms + pipeline diagrams. Write-from-memory drill',
    'Change models (Kotter/Lewin) + case applications',
    'Performance cycle + career theories + appraisal error tables',
    'Predictive analytics frameworks + HR KPIs — 1 intensive day',
    'Regression, validity/reliability — 1 day + past paper',
    'HR metrics, descriptive analytics — minimum viable review',
]
for i, (d, w, s) in enumerate(zip(data, hour_weights, strategies)):
    alloc = round(total_hrs * w, 1)
    rank_rows.append([i+1, d['name'], d['urgency_score'], d['risk'], f'{alloc}h', s])

print(tabulate(rank_rows, headers=rank_headers, tablefmt='fancy_grid'))

display(HTML(f"""
<div style='background:#d1ecf1;border-left:4px solid #17a2b8;padding:12px 16px;margin:12px 0;border-radius:4px'>
  <b>📅 Total study budget:</b> ~{total_hrs:.0f} hours over {(datetime.date(2026,6,20)-TODAY).days+1} days<br>
  Workdays: {wk} × {STUDY_HOURS_WKDAY}h = {wk*STUDY_HOURS_WKDAY}h &nbsp;|
  Saturdays: {sat} × {STUDY_HOURS_SAT}h = {sat*STUDY_HOURS_SAT}h &nbsp;|
  Sundays: {sun} × {STUDY_HOURS_SUN}h = {sun*STUDY_HOURS_SUN}h
</div>
"""))
```

```python
display(HTML(h('📈 GPA Simulator — What-If Analysis')))
display(HTML("""
<div style='background:#eaf4fb;border-left:4px solid #2980b9;padding:12px 16px;margin:10px 0;border-radius:4px'>
Edit <b>your_finals_scores</b> below as you complete each exam to see your projected GPA in real time.
</div>
"""))

# ─── EDIT THESE AFTER EACH EXAM ───
your_finals_scores = {
    'MGT388': None,   # Enter your actual final exam score out of 50 after exam
    'HRM487': None,
    'HRM485': None,
    'MGT389': None,
    'HRM478': None,
    'MGT301': None,
    'HRM479': None,
}

# ─── COMSATS GPA scale ───
def get_grade(total):
    if   total >= 90: return ('A+', 4.0)
    elif total >= 85: return ('A',  4.0)   # COMSATS: A and A+ both = 4.0
    elif total >= 80: return ('A-', 3.7)
    elif total >= 75: return ('B+', 3.3)
    elif total >= 70: return ('B',  3.0)
    elif total >= 65: return ('B-', 2.7)
    elif total >= 60: return ('C+', 2.3)
    elif total >= 55: return ('C',  2.0)
    elif total >= 50: return ('C-', 1.7)
    else:             return ('F',  0.0)

print(f"{'Subject':<35} {'Internal':>10} {'Final':>8} {'Total':>7} {'Grade':>6} {'GPA Pts':>8} {'Target':>10}")
print('─'*90)

total_gpa_pts = 0
completed = 0
all_rows = []

for code, info in INTERNALS.items():
    internal = info['internal']
    final    = your_finals_scores.get(code)
    needed   = round(85 - internal, 2)

    if final is not None:
        total   = round(internal + final, 2)
        grade, gpa_pts = get_grade(total)
        status  = '✅' if total >= 85 else ('⚠️' if total >= 70 else '❌')
        total_gpa_pts += gpa_pts
        completed += 1
    else:
        total = '—'
        grade = '—'
        gpa_pts = '—'
        status = '📌 pending'

    print(f"{status} {info['name']:<32} {internal:>10.2f} {str(final) if final else '(tbd)':>8} {str(total):>7} {grade:>6} {str(gpa_pts):>8} {needed:>10.2f}/50")

print('─'*90)
if completed > 0:
    current_gpa = total_gpa_pts / completed
    projected_gpa = total_gpa_pts / len(INTERNALS)  # assumes 0 for remaining
    print(f'\n  Subjects completed: {completed}/{len(INTERNALS)}')
    print(f'  Current GPA (completed only): {current_gpa:.3f}')
    print(f'  Projected semester GPA (if remaining = B+): {(total_gpa_pts + (len(INTERNALS)-completed)*3.3)/len(INTERNALS):.3f}')
else:
    print('\n  → Enter your final scores above after each exam to see live GPA simulation.')
    print(f'  → Target: all subjects A+ = GPA 4.000')
    print(f'  → Minimum to pass all: finals avg 41/50 across subjects')
```
